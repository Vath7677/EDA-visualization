* Adding Hue, Size, and Style: Build a sns.scatterplot() comparing molecular_weight (X-axis) against absorption_rate (Y-axis). Make it a highly detailed 4D plot by assigning hue='drug_class', style='delivery_method', and size='solubility_index'. Can you spot any visual trends regarding the size of the bubbles and their position on the Y-axis?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('synthetic_pharmaceutical.csv')
# df

sns.scatterplot(data=df, x='molecular_weight', y='absorption_rate', hue='drug_class', style='delivery_method', size='solubility_index')

plt.show()
```

* Correlation & Heatmaps: * First, isolate your numerical data by dropping the categorical text columns: num_df = df.drop(columns=['drug_class', 'delivery_method']).
    * Calculate the correlation matrix using correlation_matrix = num_df.corr().
    * Visualize it using sns.heatmap(data=correlation_matrix, annot=True, cmap='coolwarm'). Look at the scores intersecting with absorption_rate to prove your visual hypotheses from the previous steps.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('synthetic_pharmaceutical.csv')
df = df.drop(columns=['Unnamed: 0'])

num_df = df.drop(columns=['drug_class', 'delivery_method'])
correlation_matrix = num_df.corr()

sns.heatmap(data=correlation_matrix, annot=True, cmap='coolwarm')

plt.show()
```
