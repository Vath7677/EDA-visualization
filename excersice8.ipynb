```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('trading_log.csv')
# df

sns.relplot(data=df, 
            x="trade_duration_mins", 
            y="profit_pips",
            col="session",
            hue="strategy"
)

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('trading_log.csv')
# df

sns.catplot(
    data=df,
    x="strategy",
    y="profit_pips",
    row="session",
    kind="box"
)

plt.show()
```

* Joint Plots: Filter your dataframe to just look at your Breakout trades (breakouts = df[df['strategy'] == 'Breakout']).
    * Use sns.jointplot() to map trade_duration_mins on the X-axis and profit_pips on the Y-axis.
    * Observe the central scatterplot, but pay close attention to the marginal histograms on the top and right. What is the most common duration for a breakout trade?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('trading_log.csv')
df = df.drop(columns=['Unnamed: 0'])

df

# ->>>>> breakout this code line is using to filtering the "Breakout"
breakouts = df[df['strategy'] == 'Breakout']
sns.jointplot(
    data=breakouts, 
    x="trade_duration_mins", 
    y="profit_pips", 
    kind="scatter"
)

plt.show()
```

## Explain: 

* For this graph is call jointplot that have that in the plot show scatter the relationship of 2 data we using to compare and out side the plot have the bar plot using to show the data of one side. So on the graph X_line="trade_duration_mins" and Y_line="profits_pips" for the graph out side the plot is a bar plot of one data on the to of x_line call trade_duration_mins show that data on the top by using bar plot. Simalarly the top right of the plot show the data of the profit_pips. So depend on the graph jointplot the scatter showing us that the most profit that they get is a short trade_duration_mins not long as well only 0 to 25.
