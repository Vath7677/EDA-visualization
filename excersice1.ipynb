* The Anatomy / Basic Plotting: Ask them to track a week of temperatures for two cities. They must use the Object-Oriented method: fig, ax = plt.subplots(). They will plot the data using ax.plot(days, city_A_temps) and ax.plot(days, city_B_temps).

```python
import matplotlib.pyplot as plt
import seaborn as sns 

sns.set_theme(style="darkgrid")
fig, ax = plt.subplots(figsize=(8,5))

Day = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]

phnom_penh = [32, 33 , 34 ,35 ,36, 37 ,38]
kompot = [38, 37, 36, 35 ,34 ,33 ,34]

ax.plot(Day, phnom_penh, color='blue', marker='o', label='phnom_penh')
ax.plot(Day, kompot, color='green', marker='o', label='kompot')
```

* Basic Customization: Instruct them to add ax.set_title("Weekly Temperature Comparison"), ax.set_xlabel("Days of the Week"), ax.set_ylabel("Temperature (°C)"), and call ax.legend() to show which line belongs to which city.

```python
ax.set_title("Weekly Temperature Comparison") 
ax.set_xlabel("Days of the Week")
ax.set_ylabel("Temperature (°C)")
ax.legend()

plt.show()
```

* Explain each of the graphs in details

Based on the visualization of the graph, we have two plots on the same axes showing the temperature from Monday to Sunday. The blue line talks about Phnom Penh's temperature (°C). The graph shows that this week, on Monday it is only 32, on Tuesday it is 33, and the temperature grows to 38 on Sunday. But if we look at the green line that talks about Kampot province, we see that the temperature this week on Monday is 38, Tuesday is 37, Wednesday is 36, Thursday is 35, Friday is 34, and Saturday is 33, but for Sunday it goes up to 34."

# Basic coutomization(The OO Way)

```python
import matplotlib.pyplot as plt
import seaborn as sns

# using to set the color of the ax
sns.set_theme(style="darkgrid")
fig, ax = plt.subplots(figsize=(8,5))

x_data = [1,2,3,4,5]
y_data = [10,20,30,40,50]

# uing this to set what color, marker, and label using with ax.legend()
ax.plot(x_data, y_data, color='blue', marker='o', label='Sales growth')

ax.set_title("Company Sales Over 5 Years") # Add Title
ax.set_xlabel("Years")                     # Add X-axis label
ax.set_ylabel("Total Sales ($)")           # Add Y-axis label
ax.legend()     

plt.show()
```
