* FacetGrid / Relplot / Catplot: Use sns.relplot(x='weight', y='mpg', col='origin'). This will generate three separate scatterplots side-by-side (USA, Europe, Japan) so students can compare weight vs. fuel efficiency by region. Then, use sns.catplot(kind='box', x='cylinders', y='horsepower', row='origin') to create a vertical stack of boxplots.

```python
import seaborn as sns 
import pandas as pd
import matplotlib.pyplot as plt

df = sns.load_dataset('mpg')
# df

sns.relplot(
    data=df,
    x="weight",
    y="mpg",
    col="origin"
)

plt.show()

sns.catplot(
    data=df,
    x="cylinders",
    y='horsepower',
    row="origin",
    kind="box"
)

plt.show()
```

* Joint Plots: Use sns.jointplot(x='horsepower', y='mpg') to show a scatter plot in the center, with a histogram of horsepower on the top margin and a histogram of mpg on the right margin.

```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt

df = sns.load_dataset('mpg')
# df

sns.jointplot(
    data=df,
    x="horsepower",
    y="mpg",
    kind="scatter"
)

plt.show()
```
