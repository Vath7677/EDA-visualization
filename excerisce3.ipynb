## Categorical - Bar/Count

* Categorical - Bar/Count: Use sns.countplot() to visualize the treatment_arm column. Which group has the fewest patients?

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 

df = pd.read_csv('synthetic_clinical.csv')
# df

# using this to set color
colors = ["#FF2B00", "#003CFF", "#000000"]

sns.countplot(data=df, 
              x='treatment_arm', 
              hue='treatment_arm', 
              palette = colors, 
              legend=False)

plt.ylabel('Count')
plt.xlabel('Type of treatment_arm')
plt.title('Visualize the treatment_arm')


plt.show()
```

## Explain:

* This graph is call an count plot and we saw this showing us about the type of treatment_arm by count the treatment_arm in the data. After we analysis already by using the plot we see that the group has the fewest patients is Placebo that stand on the blue Bar.

# Categorical - Pie

* Categorical - Pie: Use df['patient_sex'].value_counts() to get the raw numbers, and then pass those values into plt.pie() to show the proportion of Male vs. Female patients. Include the labels=['Male', 'Female'] argument so you know which slice is which.

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 

df = pd.read_csv('synthetic_clinical.csv')

counts = df['patient_sex'].value_counts()

plt.pie(counts, 
        labels=['Male', 'Female'],   # we also can use  { labels = counts.index  } 
        autopct='%1.1f%%', 
        textprops={'color': 'black'})

plt.title('Type of Gender')

plt.show()
```

## Explain:

* After we imoort excel to use an data and plt.pie to make a pie want to know which slide is Male and Female by we using labels = ['Male', 'Female'] after we using that we see the graph show Male slide is Blue have 52.6% and Femmale is orange have 47.7% 

## Histograms

* Numerical - Histograms: Use sns.histplot() to find the distribution of the age column. Add the argument bins=15 to see how changing the bucket size changes the look of the graph.

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 

df = pd.read_csv('synthetic_clinical.csv')
# df

sns.histplot(data=df, x='age', bins=15)
plt.title('Graph of age')

plt.show()
```

## KDE

* Numerical - KDE: Use sns.kdeplot() on the blood_pressure_drop column. Notice the shape of the curve. Does it look like a symmetrical bell, or is it skewed heavily to one side?

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 

df = pd.read_csv('synthetic_clinical.csv')

sns.kdeplot(data=df, x='blood_pressure_drop', fill=True)
plt.title('blood_pressure type')

plt.show()
```

## Explain:

* sns.kdeplot() is a plot that have smooth line this using with the numerical data so after we using that we see the top tallest is range 0 to 20. For the data that have only a few is range from 80 to 100 Skewed Right. Skewed Left is a negative side have a data that few and small that why make the graph is short.  

## Combining Hist + KDE

* Combining Hist + KDE: Create a final plot using sns.histplot() for the blood_pressure_drop column, but this time set kde=True.

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 

df = pd.read_csv('synthetic_clinical.csv') 

sns.histplot(data=df, x='blood_pressure_drop', bins=20 ,kde=True)
plt.title('blood_pressure type')

plt.show()
```
