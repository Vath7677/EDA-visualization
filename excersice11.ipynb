## Dataset: sns.load_dataset('flights')

* Themes & Contexts: Apply sns.set_style('whitegrid') to give it a clean background, and sns.set_context('talk') to automatically scale up the fonts for a presentation slide.

```python
import seaborn as sns 
import matplotlib.pyplot as plt
import pandas as pd

df = sns.load_dataset("flights")

sns.set_style('whitegrid')
sns.set_context('talk')
```

* Color Palettes: Pivot the dataset into a matrix (months vs. years) and plot a heatmap. Use sns.color_palette("Blues", as_cmap=True) so the colors intuitively map from light (few passengers) to dark (many passengers).

```python
fig, ax = plt.subplots(figsize=(12, 8))
colors = sns.color_palette("Blues", as_cmap=True)

flights_pivot = df.pivot(index="month", columns='year', values='passengers')

#  { fmt='d' } we using this for making the number as an interger
sns.heatmap(data=flights_pivot, annot=True, cmap=colors, fmt='d', annot_kws={"size": 10})

plt.show()
```

* Advanced Matplotlib Formatting: Plot a line chart of average passengers per year.
    * Rotate the X-axis labels (years) by 45 degrees so they are perfectly legible.
    * Use ax.set_ylim(0, 500) to force the Y-axis to start at zero instead of floating.
    * Use ax.annotate() to draw an arrow pointing to the year 1960 with the text: "Introduction of Commercial Jets".

```python
fig, ax = plt.subplots(figsize=(10, 6))

sns.lineplot(data=df, x="year",
             y="passengers", 
             estimator='mean',
             marker="o",
             errorbar=None)

plt.xticks(rotation=45)
ax.set_ylim(0, 500)

# we using to finf an average on year '1960'
value_1960 = df[df['year']==1960]['passengers'].mean()

ax.annotate("Introduction of Commercial Jets",
            xy=(1960, value_1960),
            xytext=(1952, 450),
            arrowprops=dict(facecolor='black', shrink=0.05))

ax.set_title("Average Passengers Per Year")
plt.show()
```

* So this graph we finding about the average the passangers of the year. If we are looking on the y_line we can see the amount passangers form 0-500 and x_line is from 1950 to 1960. So on the the graph that have one line that line call line plot also have marker on that to mark the year of the averange amount passangers. After we using the tool aready we can see that in 1960 the average of the passanger is the largest. On more thing on the line plot also have black arrowsprops point on 1960.
