## Dataset: sns.load_dataset('tips')

* Num vs. Num - Scatter: Create a sns.scatterplot() with total_bill on the X-axis and tip on the Y-axis to prove that higher bills equal higher tips.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

df = sns.load_dataset('tips')
# df.head()

sns.scatterplot(data=df, x='total_bill', y='tip')

plt.show()
```

## Explain:

* This graph called scartter plot because have blue scartter on the graph. This graph is Num vs Num by we using x_line is total_bill and y_line is tip. However, if we are looking on the graph we see that the scartter plot showing us the plot that messy do not have clear line but that scartter is going up to the top. So, depending on the graph scatter plot, we can say that these two items is Positive Correlation.

* Num vs. Num - Line: Create a mock timeline of total_bill values using sns.lineplot() to practice connecting data points.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

df = sns.load_dataset('tips')
# df

sns.lineplot(data=df, y='total_bill', x='time')
plt.title("Trend of Total Bill by Time")

plt.show()
```

## Explain:

* So this graph is call lineplot that we have a line showing data go up or go down. The x_line is Time and y_line is total_bill and x_line is time. For the blue line on the graph is call an medain or mean and the shadow area is call Confidence Interval that showing us about the Indicates the degree of bias or dispersion of the data around the mean value. Depend on the graph we can conclusion that total_bill in Dinner time is better than Lucnh time.

* Num vs. Cat - Box Plots: Use sns.boxplot() with day on the X-axis and total_bill on the Y-axis to find the median bill and outliers for each day of the week.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

sns.load_dataset('tips')
# df

sns.boxplot(data=df, x='day', y='total_bill')
plt.ylabel("Total bill")

plt.show()
```

* Num vs. Cat - Bar Plots: Use sns.barplot() to show the average tip amount grouped by size (party size), letting Seaborn calculate the mean automatically.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = sns.load_dataset('tips')
# df

sns.barplot(data=df, x="size", y="tip")

plt.show()
```

## Explain:

* So this plot, the bar plot, if we are looking on the graph, we see 6 bars on the graph. The y_line call tip and x_line call size so it means we analyze the tip that people give by the side of the people. The blue bar that we see is a median or mean of tips. So we can see that the size is bigger; the tip also earns. For the black line on the bar, call it an error bar to show if the data are too different or not.
