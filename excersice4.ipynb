* Num vs. Num - Scatter: Create a sns.scatterplot() with volume_traded on the X-axis and spread_cost on the Y-axis to see if higher trading volumes correlate with wider spreads.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

df = pd.read_csv('EUR_USD_trading_data.csv')
# df

sns.set_theme(style="darkgrid")
sns.scatterplot(data=df, x='volume_traded', y='spread_cost')
# sns.regplot(data=df, x='volume_traded', y='spread_cost', line_kws={"color": "red"})

plt.title('Correlation between Trading Volume and Spread Cost')

plt.show()
```

* Num vs. Cat - Bar Plots: Use sns.barplot() to show the average spread_cost grouped by session. Let Seaborn calculate the mean to see which session is generally the most expensive to trade in.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

df = pd.read_csv('EUR_USD_trading_data.csv')
# df.head()

sns.barplot(data=df, x='session',y='spread_cost',estimator='mean')

plt.show()
```
