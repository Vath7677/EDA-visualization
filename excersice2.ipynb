## Bar/Count

* Categorical - Bar/Count: Use sns.countplot() to show how many passengers were in First, Second, and Third class (class column).

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# the sns.load_dataset is create a data
titanic = sns.load_dataset('titanic')
# titanic

plt.figure(figsize=(8, 5))

# ## ->>> using this to count only the class 
sns.countplot(data=titanic, x='pclass', hue='pclass')

plt.title('Passenger Count by Class')
plt.xlabel('Class')
plt.ylabel('Count')

plt.show()
```

## Pie plot 

* Categorical - Pie: Use plt.pie() to show the raw proportion of male vs. female passengers (sex column), discussing why the countplot might have been easier to read.

### discussing why the countplot might have been easier to read.

* The count plot might have been easier to read than the pie plot because the count plot shows a graph that allows people just looking for a minute to get the information easier without using notes or any numerical on the bar.

* But for the pie plot, it is a circle plot that we need to take a look at first and might be confusing and difficult to read if there is more than 8 data points.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

## ->>> create a Visual data
titanic = sns.load_dataset('titanic')
# titanic

counts = titanic['sex'].value_counts()

plt.pie(counts, labels=counts.index, autopct = '%1.1f%%', textprops={'color': 'black'} )
plt.title('Count Gender')
plt.show()
```

## Explain:

* Pie plot showing that have 2 data is Male and Female the biggest one is male have 64.8% using blue color on the pie but for the orange color stand for femal that have 35.2%. It is easy to look because have only 2 data in the pie plot that why it look clean.

## Numerical - Histograms

* Numerical - Histograms: Use sns.histplot() to find the age distribution (age column) of the passengers.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

## ->>> create a Visual data
titanic = sns.load_dataset('titanic')

sns.histplot(data=titanic, x='age')
plt.title("Show Age")

# sns.histplot(data=df, x='age', bins=5, kde=True )

plt.show()
```

## Explain:

* For the histograms graph showing the bar that stands for age, we see that the tallest one is between 15 and 30, and the shortest bar is range 70 to 80.

## Numerical

* Numerical - KDE: Use sns.kdeplot() to look at the density of ticket prices (fare column) to see the massive skew toward cheaper tickets.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

## ->>> create a Visual data
titanic = sns.load_dataset('titanic')
# titanic

sns.kdeplot(data=titanic, x='fare', fill=True)
plt.title('Type of pirce')

plt.show()
```

## Explain: 

* the graph that have smooth line we call { KDE plot } and this graph showing that the most tricket is range from 0 to 100. 

## Combining Hist + KDE

* Combining Hist + KDE: Overlay the KDE curve on top of a histogram for the age column using sns.histplot(data, x='age', kde=True) to see both the bins and the smooth trend.

```python
titanic = sns.load_dataset('titanic')

sns.histplot(titanic, x='age', kde=True)
plt.title('showing the EDA and Histograms')


plt.show()
```

* This graph is called KDE combined with histograms, so in the graph picture there is a smooth line on the bar, and both of them show us that they depend on the age data, so we know that the age that we have the most age between 20 and 33.
