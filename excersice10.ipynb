## Dataset: Synthetic historical Forex trading volume (Code provided below)

* Themes & Contexts: Before drawing anything, set your style to 'whitegrid' and your context to 'talk'.

```python
import pandas as  pd 
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('EUR_USD_trading_volume.csv')
df = df.drop(columns=['Unnamed: 0'])
# df

sns.set_style('whitegrid')
sns.set_context('talk')
```

* Color Palettes (The Heatmap): * Pivot the dataframe so month is the index (rows), year is the columns, and volume_billions are the values.

```python
fig, ax = plt.subplots(figsize=(12, 8))
colors = sns.color_palette("Reds", as_cmap=True)

monthly_trading = df.pivot(index='month', 
                           columns='year', 
                           values='volume_billions')
```

* Create a heatmap using sns.heatmap(). Use the "Greens" color palette (cmap="Greens"). Look at the grid: can you spot the "summer lull" where trading volume dries up?

```python
sns.heatmap(data=monthly_trading, annot=True, 
            cmap=colors, 
            fmt='.2f', 
            annot_kws={"size": 10})

plt.show()
```

* Advanced Matplotlib Formatting (The Line Chart):
    * Create a new figure and axes (fig, ax = plt.subplots(figsize=(10, 6))).
    * Group the original df by year and calculate the average volume, then plot it as a line chart using ax.plot().
    * Set the Y-axis limit to start at 0 using ax.set_ylim().
    * Use ax.annotate() to draw an arrow pointing to the massive spike in the year 2020. Add the text: "2020 Market Volatility".

```python
import pandas as  pd 
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('EUR_USD_trading_volume.csv')
df = df.drop(columns=['Unnamed: 0'])
# df

sns.set_style('whitegrid')
sns.set_context('talk')

fig, ax = plt.subplots(figsize=(10, 6))

yearly_avg = df.groupby('year')['volume_billions'].mean()
ax.plot(yearly_avg.index, 
        yearly_avg.values,
        marker='o',
        color='royalblue')

ax.annotate("2020 Market Volatility", 
            xy=(2020, yearly_avg[2020]),  
            xytext=(2018, yearly_avg.max()), 
            arrowprops=dict(facecolor='black', shrink=0.05))

ax.set_ylim(bottom=0)

ax.set_title('Average Annual Trading Volume')
ax.set_xlabel('Year')
ax.set_ylabel('Average Volume (Billions)')
ax.grid(True)

plt.show()
```
